```python
# 导入包
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import dlib
# %matplotlib inline
plt.rcParams['figure.dpi'] = 200
```

```python
# 获取人脸的68个关键点
# 人脸检测模型
hog_face_detector = dlib.get_frontal_face_detector()
```

```python
# 关键点检测模型
shape_detector = dlib.shape_predictor('./weights/shape_predictor_68_face_landmarks.dat')
```

```python
# 读取一张侧视图
img = cv2.imread('./images/faces2.jpg')
img.shape
# plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
# 检测人脸
detections = hog_face_detector(img,1)# 缩放因子
```

```python
for face in detections:
    # 人脸框坐标
    l,t,r,b = face.left(),face.top(),face.right(),face.bottom()
    # 获取68个人脸关键点
    points = shape_detector(img,face)
    # 绘制关键点
    for point in points.parts():
        cv2.circle(img,(point.x,point.y),2,(0,255,0),1)
    # 绘制矩形框
    cv2.rectangle(img,(l,t),(r,b),(0,255,0),2)
    
```

```python
# 面部特征描述符
# 获取人脸的68个关键点
# 人脸检测模型
hog_face_detector = dlib.get_frontal_face_detector()
# 关键点检测模型
shape_detector = dlib.shape_predictor('./weights/shape_predictor_68_face_landmarks.dat')
# resnet模型
face_decriptor_extractor = dlib.face_recognition_model_v1('./weights/dlib_face_recognition_resnet_model_v1.dat')
```

```python
# 提取单张图片的特征描述符，label
def getFaceFeatLabel(fileName):
    # 获取人脸label
    label_id = int(fileName.split('/')[-1].split('.')[0].split('subject')[-1])
    # 读取图片
    cap = cv2.VideoCapture(fileName)
    ret,img = cap.read()
    # 转为rgb
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    # 人脸检测
    detections = hog_face_detector(img,1)
    face_decriptor = None
    for face in detections:
    # 获取关键点
        points = shape_detector(img,face)
        # 获取特征描述符
        face_decriptor = face_decriptor_extractor.compute_face_descriptor(img,points)
        # 转为numpy二维数组
        face_decriptor = [f for f in face_decriptor]
        face_decriptor= np.asarray(face_decriptor,dtype=np.float64)
        face_decriptor= np.reshape(face_decriptor,(1,-1))
    return label_id,face_decriptor    
```

```python
# 遍历train文件夹，对所有图片同样处理
# 拼接成大的list
import glob
file_list = glob.glob('./yalefaces/train/*')
# 构造两个空列表
label_list = []
feature_list = None
name_list = {}
index = 0
for train_file in file_list:
    # 获取每一张图片得对应信息
    label,feat = getFaceFeatLabel(train_file)
    # 过滤数据
    if feat is not None: 
        # 文件名列表
        name_list[index] = train_file
          # 标签列表
        label_list.append(label)
        if feature_list is None:
            feature_list = feat
        else:
            # 特征列表
            feature_list = np.concatenate((feature_list,feat),axis=0)
        index += 1
```

```python
# 评估测试数据集

file_list = glob.glob('./yalefaces/test/*')
# 构造两个空列表
predict_list= []
label_list = []
# 判断的距离阈值
threshold = 0.5
for test_file in file_list:
    # 获取每一张图片得对应信息
    label,feat = getFaceFeatLabel(test_file)

    # 读取图片
    cap = cv2.VideoCapture(test_file)
    ret,img = cap.read()
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
    # 过滤数据
    if feat is not None: 
        # 计算距离
        distance = np.linalg.norm((feat-feature_list),axis=1)
        # 寻找最短距离
        min_index = np.argmin(distance)
        min_distance = distance[min_index]

        if min_distance < threshold:
            # 同一个人
            predict_id = int(name_list[min_index].split('/')[-1].split('.')[0].split('subject')[-1])
        else:
            predict_id = -1
        predict_list.append(predict_id)
        label_list.append(label)

        cv2.putText(img,'True:'+str(label),(10,30),cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(0,255,0))
        cv2.putText(img,'Pred:'+str(predict_id),(10,50),cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(0,255,0))
        cv2.putText(img,'Distance:'+str(min_distance),(10,70),cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(0,255,0))

        #显示多张图片
        plt.figure()
        plt.imshow(img)
        
```
